# Visualization of Images per Every Year in SAPO.PT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/Users/emilis/Desktop/DATA/FCD_SAPO/ImageCount.csv")

data.columns = data.columns.str.strip()

data['year'] = data['year'].astype(int)

plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='image_count', data=data, marker='o', 
label='Number of Images', color='b')

plt.title("Number of Images Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Images")
plt.legend()

plt.xticks(ticks=range(data['year'].min(), data['year'].max() + 1), rotation=45)

plt.tight_layout()
plt.show()
```

# Website Size and Most Used Font over Years:

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_graphs(csv_file):

    df = pd.read_csv(csv_file)

    plt.figure(figsize=(10, 6))
    plt.bar(df['year'], df['average_font_size_px'], color='skyblue')
    plt.xlabel('Year')
    plt.ylabel('Average Font Size (px)')
    plt.title('Average Font Size by Year')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 6))
    most_used_font_counts = df.groupby('most_used_font')['font_usage_count'].sum().sort_values(ascending=False)

    most_used_font_counts.plot(kind='bar', color='lightgreen')
    plt.xlabel('Font')
    plt.ylabel('Total Font Usage Count')
    plt.title('Total Font Usage Count by Font')
    plt.tight_layout()
    plt.show()

plot_graphs('TypographyReport.csv')
```
